```python
student_names = []

print("\n[1] Add Students")

for i in range(3):
    name = input(f"Enter student name #{i+1}: ")
    student_names.append(name)

print("\nCurrent List:", student_names)

print("\n[2] Find a Student")

search_name = input("Enter a name to search: ")

if search_name in student_names:
    position = student_names.index(search_name)
    print(f"Found '{search_name}' at index {position}")
else:
    print(f"'{search_name}' not found")

print("\n[3] Remove a Student")

remove_name = input("Enter a name to remove: ")

if remove_name in student_names:
    student_names.remove(remove_name)
    print(f"'{remove_name}' has been removed")
else:
    print(f"'{remove_name}' not found")

print("Updated List:", student_names)

print("\n[4] Sort Students")

student_names.sort()

print("Sorted List:", student_names)

print("\n[5] Count Students")

print("Total Students:", len(student_names))



while True:

    print("\n========== MENU ==========")
    print("1. Add Student")
    print("2. Remove Student")
    print("3. Find Student")
    print("4. Sort Students")
    print("5. View Students")
    print("6. Count Students")
    print("7. Exit")

    choice = input("Choose menu (1-7): ")

    if choice == "1":

        name = input("Student name: ")
        student_names.append(name)

        print("Added successfully.")

    elif choice == "2":

        name = input("Name to remove: ")

        if name in student_names:
            student_names.remove(name)
            print("Removed successfully.")
        else:
            print("Student not found.")

    elif choice == "3":

        name = input("Name to search: ")

        if name in student_names:
            print("Found at index",
                  student_names.index(name))
        else:
            print("Student not found.")

    elif choice == "4":

        student_names.sort()
        print("Students sorted.")

    elif choice == "5":

        if len(student_names) == 0:
            print("No students.")
        else:
            print("\nStudent List")
            for i in range(len(student_names)):
                print(f"{i}. {student_names[i]}")

    elif choice == "6":

        print("Total Students:", len(student_names))

    elif choice == "7":

        print("Exit Program")
        break

    else:
        print("Invalid Menu! Please choose again.")
```
