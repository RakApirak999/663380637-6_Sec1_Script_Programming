```python
todo_list = []

while True:
    print("\n===== To-Do List =====")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Complete Task")
    print("4. Exit")

    choice = input("Choose: ")

    if choice == "1":
        task = input("Enter task: ")
        todo_list.append(task)

    elif choice == "2":
        if len(todo_list) == 0:
            print("No tasks.")
        else:
            for i, task in enumerate(todo_list):
                print(f"{i+1}. {task}")

    elif choice == "3":
        task = input("Enter task name to complete: ")
        if task in todo_list:
            todo_list.remove(task)
            print("Task completed.")
        else:
            print("Task not found.")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice")
```
