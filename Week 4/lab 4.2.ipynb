```python
import math

point1 = (3, 4)
point2 = (6, 8)

print("\n[1] Access Coordinates")

print(f"Point 1: x = {point1[0]}, y = {point1[1]}")
print(f"Point 2: x = {point2[0]}, y = {point2[1]}")

print("\n[2] Attempt Modification (Immutability Demonstration)")

try:
    point1[0] = 5
except TypeError as e:
    print("Error trying to modify tuple:")
    print(e)

print("\n[3] Calculate Distance")

distance = math.sqrt(
    (point2[0] - point1[0]) ** 2 +
    (point2[1] - point1[1]) ** 2
)

print(f"Distance between {point1} and {point2} = {distance}")


x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))

x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))

point1 = (x1, y1)
point2 = (x2, y2)

print("\nNew Points")
print(f"Point 1 = {point1}")
print(f"Point 2 = {point2}")

distance = math.sqrt(
    (point2[0] - point1[0]) ** 2 +
    (point2[1] - point1[1]) ** 2
)

print(f"Distance = {distance}")
```
