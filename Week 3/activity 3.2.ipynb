```python
for i in range(5):
    for j in range(5):
        print("*", end="")
    print()
```

```python
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()
```

```python
for i in range(5, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```

```python
height = 5

for row in range(1, height + 1):

    for space in range(height - row):
        print(" ", end="")

    for star in range(2 * row - 1):
        print("*", end="")

    print()
```
