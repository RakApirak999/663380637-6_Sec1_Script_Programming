Activity 2.2:

1. เนื้อเรื่องเริ่มต้น (Story): คุณตื่นขึ้นมากลางห้องโถงของปราสาทร้างอันมืดมิด ด้านหน้าของคุณมีประตูอยู่ 2 บาน คือ "ประตูสีแดง" และ "ประตูสีน้ำเงิน"
2. การรับข้อมูลจากผู้เล่น (INPUT): ใช้คำสั่ง INPUT() เพื่อรับทางเลือกจากผู้เล่น โดยให้พิมพ์คำว่า "แดง" หรือ "น้ำเงิน"
3. ลำดับการทำงานด้วยเงื่อนไข (IF-ELIF-ELSE):
- IF: ผู้เล่นกรอกข้อมูลเป็นคำว่า "แดง" * เหตุการณ์: คุณเปิดประตูสีแดงเข้าไปพบกับมังกรพ่นไฟที่กำลังนอนหลับอยู่ แต่คุณดันเดินเหยียบกิ่งไม้เสียงดังจนมันตื่นขึ้นมา * ผลลัพธ์ (Output): "มังกรตื่นขึ้นมาพ่นไฟใส่คุณ! Game Over!"
- ELIF: ผู้เล่นกรอกข้อมูลเป็นคำว่า "น้ำเงิน" * เหตุการณ์: คุณเปิดประตูสีน้ำเงินเข้าไป พบกับหีบสมบัติทองคำวางอยู่กลางห้องอย่างปลอดภัย ไม่มีอันตรายใดๆ * ผลลัพธ์ (Output): "ยินดีด้วย! คุณพบขุมทรัพย์มหาศาลและรอดชีวิตออกจากปราสาทได้! You Win!"
- ELSE: ผู้เล่นกรอกข้อมูลอื่นๆ ที่ไม่ใช่ "แดง" หรือ "น้ำเงิน" * เหตุการณ์: คุณยืนลังเลไม่ยอมเลือกประตูบานใดเลยเป็นเวลานานเกินไป * ผลลัพธ์ (Output): "คุณถูกวิญญาณในปราสาทเข้าสิงเพราะความลังเล! Game Over!"
